# Preprocessing in R
Ambient RNA contamination removal and Doublet finding

https://www.10xgenomics.com/datasets/160k_DTC_Matched_PBMC_MultiPro_Human_Discovery_Panel

```r
set.seed(7)
library(SoupX)
library(Seurat)
library(scDblFinder)
library(zellkonverter)
library(hdf5r)
```

```r
# BiocManager::install("zellkonverter")
```

```r
input_dir <- file.path("data", "pbmc_2control_lc")

# dirs <- list.dirs(input_dir, full.names=TRUE, recursive=FALSE)
samples <- list()
clusters <- list()
for (sample in c("PBMC_Control1", "PBMC_Control2", "Lung_Cancer_PBMC")) {
  name <- sample
  print(paste("Loading data for sample", name))

  tod = Seurat::Read10X_h5(file.path(input_dir, paste0(sample, "_count_sample_raw_feature_bc_matrix.h5")))[["Gene Expression"]]
  toc = Seurat::Read10X_h5(file.path(input_dir, paste0(sample, "_count_sample_filtered_feature_bc_matrix.h5")))[["Gene Expression"]]
  filt_genes <- rownames(toc)
  tod <- tod[filt_genes, ]

  seurat <- CreateSeuratObject(counts=toc)
  seurat <- SCTransform(seurat)
  seurat <- RunPCA(seurat)
  seurat <- RunUMAP(seurat, dims=1:30)
  seurat <- FindNeighbors(seurat, dims=1:30)
  seurat <- FindClusters(seurat)
  meta <- seurat@meta.data
  sample_clusters <- setNames(meta$seurat_clusters, rownames(meta))

  sc <- SoupChannel(tod, toc)
  sc <- setClusters(sc, sample_clusters)
  sc <- setDR(sc, seurat@reductions$umap@cell.embeddings)
  print(head(meta))
  sc <- autoEstCont(sc)
  out <- adjustCounts(sc, roundToInt=TRUE)

  samples[[name]] <- out
  clusters[[name]] <- sample_clusters
}
```

```r
for (s in names(samples)) {
    sce <- SingleCellExperiment(assays = list(counts = samples[[s]]))
    sce <- scDblFinder(sce, clusters=clusters[[s]])
    samples[[s]] <- sce
}
```

```r
output_dir <- data_home
for (s in names(samples)) {
    zellkonverter::writeH5AD(samples[[s]], file.path(output_dir, paste0(s, ".h5ad")))
}
```
